### 로지스틱 회귀(LOgistic Regression) 분류

```python
import numpy as np
import matplotlib.pyplot as plt
```

- sigmoid 함수

```python
def sigmoid(x,a=1,b=0):
    return(1. / (1 + np.exp(-a * (x-b))))
```

```python
xs = np.linspace(-5,5,1000)
ys = sigmoid(xs)
plt.plot(xs,ys, label = 'sigmoid')
plt.plot(xs,ys*(1-ys), label = 'derivative')
plt.title('Sigmoid fucntion')
plt.yticks([0, 0.5, 1])
plt.grid(), plt.legend()
plt.show()
```

```python
# a 값에 따른 변화
y3 = sigmoid(xs, a=3)
y_half = sigmoid(xs,a=0.5)
plt.plot(xs,ys, label = 'a=1')
plt.plot(xs,y3, label = 'a = 3') 
plt.plot(xs,y_half, label = 'a = 0.5')
plt.title('Sigmoid fucntion')
plt.yticks([0, 0.5, 1])
plt.grid(), plt.legend()
plt.show()
```

```python
# b 값에 따른 변화
y_plus2 = sigmoid(xs,b = 2)
y_minus2 = sigmoid(xs, b=-2)
plt.plot(xs,ys, label = 'b=0')
plt.plot(xs,y_plus2, label = 'b = 2')
plt.plot(xs,y_minus2, label = 'b = -2')
plt.title('Sigmoid fucntion')
plt.yticks([0, 0.5, 1])
plt.grid(), plt.legend()
plt.show()
```

- 이진 분류

```python
import pandas as pd
pima = pd.read_csv('../01.Intro/pima-indians-diabetes.csv', skiprows=9, header=None)
pima.columns = ['P','G','BP','S','I','BMI','D','Age','Target']
pima.head()
```

```python
y = pima.Target.values
X = pima.iloc[:,:-1].values
```

```python
# 로지스틱 회귀를 사용하려면 데이터를 표준화 또는 정규화 해주어야 한다.
from sklearn.preprocessing import StandardScaler
X_std = StandardScaler().fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X_std, y, stratify= y, test_size=0.2, random_state=2023
)
```

```python
from sklearn.linear_model import LogisticRegression
lrc = LogisticRegression(random_state=2023)
lrc.get_params()
```

```python
lrc.fit(X_train,y_train)
lrc.score(X_test,y_test)
```

```python
# 선형회귀의 가중치(계수)
lrc.coef_
```

```python
# 선형회귀의 bias(절편)
lrc.intercept_
```

```python
X_test[0],y_test[0]
```

```python
# X_test[0]가 1이될 확률, sigmoid 함수 적용 전 값
val = np.dot(lrc.coef_, X_test[0]) + lrc.intercept_
val[0]
```

```python
# sigmoid 함수 적용
sigmoid(val)[0]
```

```python
lrc.predict_proba(X_test[:1])
```

- 다중 분류

```python
from sklearn.datasets import load_iris
iris = load_iris()
iris_std = StandardScaler().fit_transform(iris.data)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    iris_std, iris.target, stratify= iris.target, test_size=0.2, random_state=2023
)
```

```python
lrc = LogisticRegression(random_state=2023)
lrc.fit(X_train,y_train)
lrc.get_params()
lrc.score(X_test,y_test)
```

```python
lrc.predict_proba(X_test[:1])
```

```python
lrc.coef_, lrc.intercept_
```

```python
val = np.dot(lrc.coef_, X_test[0]) + lrc.intercept_
val
```

```python
sigmoid(val)
```

```python
def softmax(x):
    return np.exp(x) / np.sum(np.exp(x))
```

```python

softmax(np.array([0.1,1,2]))
```

```python
# 다중분류에서 predict_proba() 값은 softmax() 적용 값
softmax(val)
```
