### K-Means Algorithm =Iris

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python

iris = load_iris()
iris_std = StandardScaler().fit_transform(iris.data)
pca = PCA(2)
iris_pca = pca.fit_transform(iris_std)
```

#### 1. 군집화(Clustering)

```python
from sklearn.cluster import KMeans
KMeans().get_params()
```

```python
Kmeans = KMeans(n_clusters=3,verbose=1,random_state=2023, n_init='auto')
Kmeans.fit(iris_std)        # 비지도 학습이므로 y값제공 x
```

```python
# 학습 결과
Kmeans.labels_
```

```python
# G학습 결과 넘버리은 집단이 큰 순서부터 부여 
np.unique(Kmeans.labels_, return_counts=True)
```

```python
df = pd.DataFrame(iris_pca, columns=['PC1','PC2'])
df['target'] = iris.target
df['cluster'] =Kmeans.labels_
df.head(54).tail()
```

```python
# 데이터 분포
df.groupby(['target','cluster'])[['PC1']].count()
```

```python
markers = ['^','s','o']         # setosa, versicolor, verginica
for i, marker in enumerate(markers):
    x_data = df[df.cluster == i]['PC1']
    y_data = df[df.cluster == i]['PC2']
    plt.scatter(x_data,y_data,marker=marker,label='cluster'+str(i))
    plt.scatter(x_data.mean(),y_data.mean(), marker='*', c='red', s=100)
plt.legend()
plt.xlabel('PC1'), plt.ylabel('PC2')
plt.title("Clustered result")
plt.show()
```

- PCA Original Data와 비교

```python
plt.figure(figsize=(10,4))
for k, column in enumerate(['target','cluster']):
    plt.subplot(1,2,k+1)
    for i, marker in enumerate(markers):
        x_data = df[df[column] == i]['PC1']
        y_data = df[df[column] == i]['PC2']
        if k == 0: 
            plt.scatter(x_data,y_data,marker=marker,label=iris.target_names[i])
        else:
            plt.scatter(x_data,y_data,marker=marker,label='cluster'+str(i))
            plt.scatter(x_data.mean(),y_data.mean(), marker='*', c='red', s=100)
    plt.legend()
    plt.xlabel('PC1'), plt.ylabel('PC2')
    if k==0:
        plt.title("Original Data")
    else:
        plt.title("Clustered Data")

plt.show()
```
