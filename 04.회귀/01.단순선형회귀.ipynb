### 단순 선형회귀

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### 아버지와 아들의 키- Galton

```python
df= pd.read_csv('data/Galton.txt', sep='\t')
df.head()
```

- 아버지와 아들의 키만 발췌
- 칼럼명을 Father, Son으로 변경
- 단위(inch-->cm) 변경

```python
# Filtering
df = df[df.Gender == 'M']
df.head()
```

```python
# Selectiondf
df = df[['Father','Height']]
# 컬러명 변경
df.columns = ['Father','Son']
```

```python
# 단위 변경
df = df * 2.54
df.head()
```

- 산점도 구하기

```python
plt.figure(figsize=(5,5))
plt.scatter(df.Father,df.Son)
plt.xlabel("Fater's height (cm)")
plt.ylabel("Son's height (cm)")
plt.grid()
plt.show();
```

- 회귀선 구하기
- np.linalg.lstsq()         #최소 자승법(List square)
- np.polyfit()              

```python
weight, bias = np.polyfit(df.Father,df.Son,1)   # X, Y, degree(차수)
weight, bias
```

```python
# 2차식 회귀식
np.polyfit(df.Father, df.Son, 2)
```

```python
df.Father.max(), df.Father.min()
```

```python
# 회귀선(직선)
xs = np.array([157,200])
ys = xs * weight + bias
ys
```

```python
plt.figure(figsize=(5,5))
plt.scatter(df.Father,df.Son)
plt.plot(xs,ys,lw=3,c='red')
plt.xlabel("Fater's height (cm)")
plt.ylabel("Son's height (cm)")
plt.grid()
plt.show();
```

```python
plt.figure(figsize=(5,5))
sns.regplot(x='Father',y='Son', data=df)
plt.show()
```

- Scikit-learn으로 회귀식 구하기

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.get_params()
```

```python
# X 데이터의 차원을 2차원으로 바꾸어 줘야함.
X = df.Father.values.reshape(-1,1)
y = df.Son.values
X.shape, y.shape
```

```python
# 학습
lr.fit(X, y)
```

```python
# weight, bias
lr.coef_, lr.intercept_
# np.polyfit()으로 구한값:(array([0.44774792]), 97.17763700739333)
```

=평가 -R squared

```python
lr.score(X,y)   # 1-sum of squear error / sum of square total
```

```python
from sklearn.metrics import r2_score
pred = lr.predict(X)
r2_score(y, pred)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y, pred)
```
