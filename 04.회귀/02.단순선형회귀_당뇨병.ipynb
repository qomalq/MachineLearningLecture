### 단순 선형회귀 - 당뇨병 사례

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

당뇨병 데이터셋 탐색

```python

from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
```

```python
df = pd.DataFrame(diabetes.data,columns=diabetes.feature_names)
df['target'] = diabetes.target
df.head()
```

```python
df.describe()   #정규화 N(0,0.0476)로 정규화 되어있음
```

```python
print(diabetes)
```

- 10개의 feature와 target의 상관관계도

```python
fig, axs =  plt.subplots(2,5, figsize=(16,18))
for i,feature in enumerate(df.columns[:-1]):
    row, col = i // 5, i%5
    sns.regplot(x=feature, y='target', data=df, ax=axs[row][col])
```

- 10개 feature vs. target

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
# 각 feature 와 target의 선형회귀를 했을 때의 R-squared value
for feature in df.columns[:-1]:
    X = df[feature].values.reshape(-1,1)
    lr.fit(X, diabetes.target)
    score = lr.score(X, diabetes.target)
    print(f'{feature}:{score:.4f}')
```

##### BMI vs target

- 훈련/테스트 데이터셋 분리

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    df.bmi.values.reshape(-1,1), diabetes.target, test_size=0.1, random_state=2023
)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
# 가중치(계수), bias(절편)
lr.coef_, lr.intercept_
```

```python
# R squared
lr.score(X_test,y_test)
```

```python
# R squared, MSE
from sklearn.metrics import r2_score, mean_squared_error
pred = lr.predict(X_test)
r2_score(y_test,pred), mean_squared_error(y_test,pred)
```

- 시각화

```python
plt.figure(figsize=(6,4))
plt.scatter(X_train,y_train)
plt.grid()
plt.xlabel('BMI'), plt.ylabel('target')
plt.title('BMI vs Target')
```

```python
xs = np.array([-0.1,0.189])
ys = np.dot(lr.coef_[0], xs) + lr.intercept_
ys
```

```python
plt.figure(figsize=(6,4))
plt.scatter(X_train,y_train)
plt.plot(xs,ys, 'r-', lw=3)
plt.scatter(X_test,y_test,marker='^', label='test dataset')
plt.grid(),plt.legend()
plt.xlabel('BMI'), plt.ylabel('target')
plt.title('BMI vs Target')
```

##### BP vs target

```python
X_train2,X_test2, y_train2, y_test2 = train_test_split(
    df.bp.values.reshape(-1,1), diabetes.target, test_size=0.1, random_state=2023
)
lr = lr.fit(X_train2,y_train2)
pred2 = lr.predict(X_test2)
```

```python
r2_score(y_test2,pred2), mean_squared_error(y_test2,pred2)
```
